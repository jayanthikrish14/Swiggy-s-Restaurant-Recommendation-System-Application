```python
import pandas as pd

cleaned_df = pd.read_csv(r'D:\AIML\Swiggy RRS Project\Data\cleaned_data.csv')
cleaned_df
```

```python
# Drop the columns not required for pre processing
cleaned_df.drop(columns=['name','address','rating_count'], axis=1, inplace=True)
cleaned_df
```

```python
# Convert the rating column to numeric to encode
cleaned_df['rating'] = pd.to_numeric(cleaned_df['rating'], errors='coerce')
cleaned_df['rating'].isna().sum()
```

```python
# Fill the NaN values in rating column by it's mean
cleaned_df['rating'] = cleaned_df['rating'].fillna(cleaned_df['rating'].mean())
cleaned_df['rating'] .value_counts()
```

```python
# Convert the cost column to numeric to encode
cleaned_df['cost'] = cleaned_df['cost'].str.replace('₹ ','').astype(int)
cleaned_df['cost']
```

```python
cleaned_df.info()
```

```python
# Make the cuisine column an iterable of iterables
cleaned_df['cuisine'] = cleaned_df['cuisine'].str.split(',')
cleaned_df
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Instantiate MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Encode the cuisine column
cuisine_encoded = mlb.fit_transform(cleaned_df['cuisine'])

# create a new DataFrame for the encoded cuisine column
cuisine_encoded_df = pd.DataFrame(cuisine_encoded, columns=mlb.classes_, index=cleaned_df.index)

print(cuisine_encoded_df)
```

```python
import pickle
# Create a pickle file for the encoded column cuisine
with open(r'D:\AIML\Swiggy RRS Project\Data\cuisine_encoder.pkl','wb+') as ef:
    pickle.dump(cuisine_encoded_df,ef)
```

```python
# Create a new DataFrame with only city column to encode
city_df = cleaned_df.copy(deep=True)
city_df.drop(columns=['rating','cost','cuisine'], axis=1, inplace=True)
print(city_df)
```

```python
print(cleaned_df)
```

```python
# Encode the city column
city_encoded_df = pd.get_dummies(city_df['city'], dtype='int')
```

```python
print(city_encoded_df)
```

```python
import pickle

# Create a pickle file for the encoded column city
with open(r'D:\AIML\Swiggy RRS Project\Data\city_encoder.pkl','wb+') as ef:
    pickle.dump(city_encoded_df,ef)
```

```python
# Concatenate encoded cuisine columns with the cleaned DataFrame
encoded_df = pd.concat([cleaned_df.drop('cuisine', axis=1), cuisine_encoded_df], axis=1)
print(encoded_df)
```

```python
# Concatenate encoded city columns with the encoded cuisine colums
encoded_df = pd.concat([encoded_df.drop('city', axis=1), city_encoded_df], axis=1)
print(encoded_df)
```

```python
# Save the encoded data in encoded_data.csv
encoded_df.to_csv(r'D:\AIML\Swiggy RRS Project\Data\encoded_data.csv', index=False)
```
