```python
import pandas as pd

path = r"D:\AIML\Swiggy RRS Project\Data\swiggy.csv"
# Read data from the "Swiggy.csv" file and store it in a DataFrame named 'df'
df = pd.read_csv(path).head(100)
df
```

```python
# Drop irrelevant columns
df.drop(columns=['id', 'lic_no', 'link', 'menu'], axis=1, inplace=True)
df
```

```python
df.isna().sum()
# Drop null values
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Drop duplicate rows if any
df.drop_duplicates(inplace=True)
df
```

```python
# Save the cleaned data in cleaned_data.csv
df.to_csv(r'D:\AIML\Swiggy RRS Project\Data\cleaned_data.csv', index=False)
```
